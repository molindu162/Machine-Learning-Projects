```python
train_set_x = train_set_x_flatten / 255.
test_set_x = test_set_x_flatten / 255.
```

```python
model = Model()
```

```python
w, b = model.initialize_params(train_set_x_flatten.shape[0])
```

```python
params, derivatives, costs = model.optimize(w, b, train_set_x, cat_train_y, 0.005, 2000, True)
```

```python
model.fit(train_set_x, cat_train_y)
```
